# LangChain - Document Loader Integration

```python
from langchain.document_loaders import HNLoader
from langchain.prompts import ChatPromptTemplate, HumanMessagePromptTemplate
from langchain.chat_models import ChatOpenAI
```

```python
loader = HNLoader("https://news.ycombinator.com/item?id=36697119")
```

```python
data = loader.load()
```

```python
print(data[0].page_content)
```

## Chat Model

```python
api_key = open('../../api_key.txt').read()
```

```python
model = ChatOpenAI(openai_api_key=api_key)
```

```python
human_template = "Please give me a short summary of the following Hacker News comments: \n{comment}"
human_message_prompt = HumanMessagePromptTemplate.from_template(human_template)
```

```python
chat_prompt = ChatPromptTemplate.from_messages([human_message_prompt])
```

```python
model_request = chat_prompt.format_prompt(comment=data[0].page_content).to_messages()
```

```python
result = model(model_request)
```
